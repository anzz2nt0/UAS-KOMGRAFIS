```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("stock.csv")
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))
axes[0].bar(df.index, df['Purchase Price'], color='blue', edgecolor='black')
axes[0].set_title('Purchase Price Distribution', fontsize=14)
axes[0].set_xlabel('Index', fontsize=12)
axes[0].set_ylabel('Purchase Price', fontsize=12)
axes[1].bar(df.index, df['Current Price'], color='orange', edgecolor='black')
axes[1].set_title('Current Price Distribution', fontsize=14)
axes[1].set_xlabel('Index', fontsize=12)
axes[1].set_ylabel('Current Price', fontsize=12)
axes[2].bar(df.index, df['Gain/Loss'], color='green', edgecolor='black')
axes[2].set_title('Gain/Loss Distribution', fontsize=14)
axes[2].set_xlabel('Index', fontsize=12)
axes[2].set_ylabel('Gain/Loss', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("stock.csv")
data = df.sort_values(by="Dividend/share", ascending=False).head(5)
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Stock Name', y='Dividend/share', data=data, palette='YlGnBu')
plt.xlabel('Stock Names', fontsize=12, weight='bold')
plt.ylabel('Dividend per Share', fontsize=12, weight='bold')
plt.title('Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("stock.csv")
top_market = df.sort_values('Market Value', ascending=False).head(5)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x='Stock Name', y='Market Value', data=top_market, palette='viridis')
plt.xlabel('Stock Name', fontsize=12, weight='bold')
plt.ylabel('Market Value', fontsize=12, weight='bold')
plt.title('Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("stock.csv")
sorted_data = df.sort_values('Annual Yield', ascending=False).head(7)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x='Stock Name', y='Annual Yield', data=sorted_data, palette='muted')
plt.xlabel('Stock Names', fontsize=12, weight='bold')
plt.ylabel('Annual Yield', fontsize=12, weight='bold')
plt.title('Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
